```python
# Importamos libreria: Urllib y BeautifulSoup
import urllib
from bs4 import BeautifulSoup

# Variable que almacena la pagina que queremos trabajar
direccion = 'https://itse.gob.ar/view/i1.php#oferta'
# Hacemos el request
codigo = urllib.request.urlopen(direccion)
# Variable de extraccion
bea = BeautifulSoup(codigo)
# Variable de etiqueta
p = bea('p') 

texto = '' 
for x in p:
    if (len(x.attrs)) == 0: 
        a= x.contents[0] 
        texto = str(texto) +  str(a)

# Creamos un diccionario
almacenar_contar = dict() 
# Normalizo las palabras
texto = texto.replace(',','').replace('.','').replace(':','').replace('?','').replace('(','').replace(')','') 
# Dejamos todas las palabras a minusculas y cortamos 
letras = texto.lower().split() 
for x in letras:
    if len(x)>3: 
        # Almacenamos en el diccionario palabras con mas de 3 letras
        almacenar_contar[x] = almacenar_contar.get(x,0) + 1 

# Importamos libreria: Pandas
import pandas as pd

# Guardamos las palabras en un dataframe
pd.DataFrame(list(almacenar_contar.items()), columns=['palabra', 'contador']).sort_values('contador', ascending=False )
```
