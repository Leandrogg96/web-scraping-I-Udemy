```python
# Importamos libreria: Urllib y BeautifulSoup
import urllib
from bs4 import BeautifulSoup

# Pedimos el ingreso de la pagina que queremos analizar
pagina_web = input('Enter - ')
# Hacemos el request
codigo = urllib.request.urlopen(pagina_web)
# Extraemos las etiquetas
bea = BeautifulSoup(codigo)

# Elegimos la etiqueta que queramos extraer
etiqueta = bea('a')
for x in etiqueta:
    # Imprimimos por pantalla tanto la etiqueta de interes como algunos de sus atributos
    print('\n')
    print('Etiqueta "a" completa:', x)
    print('Dirección:', x.get('href'))
    print('Contenido:', x.contents)
    print('Atributo:', x.attrs)
```
